```python

import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from enum import Enum, auto
from random import uniform, random
from math import cos, sin, pi, inf, sqrt
import os

this_dir = "C:\\Users\\na366\\Desktop\\Python\\SIR"
radius = 0.075
dt = 0.1        # 時間刻み.dt = 0.01
half = 1        # half = 5
speed = 1
time_to_recover = 0.5    # 回復time_to_recover = 2（回復率 γ の逆数）.
buff = 1        # バッファ. buff = 300

fig = plt.figure(figsize=plt.figaspect(1/1))
fig.set_facecolor('black')


#               auto_run_class_(Enum)

class State(Enum):
    Susceptible = auto()
    Infected = auto()
    R1=Recovered = auto()
    print("  30_Stae.S,I,R=",Susceptible, int(Infected), int(R1),"\n")
    
class Person:
    
    def __init__( self, co, state, standstill ):
        
        self.co = co                        # 現在位置
        
        # 現在位置の記録で、過去位置保存？
        self.coords = np.array([self.co])    #  ([]) ⇒ [] 配列
        
        print("  41_Person_np.array([self.co])=",self.coords,"\n")
        
        #    [[0,0]] 2重[[]] は？
        
        self.arg = uniform(-pi, pi)    # 角度、乱数区間
        self.standstill = standstill    # 停止standstill
        
        if self.standstill:
            vel = np.array([0.0, 0.0])
        else:
            vel = np.array([speed * cos(self.arg), speed * sin(self.arg)])

        self.vel = vel                # xy速度[x_0,y_0]、停止
        self.state = state            # S,I,R
        self.infection_start = inf    # 感染スタート、無限？

    def move( self ):
        
        # (self.standsrill =Ture =停止)→ (動かす)
        #                    line_231_call
        if self.standstill:
            self.vel = np.array([speed * cos(self.arg), speed * sin(self.arg)])
            
            
# Person_instance変数は、7要素
#  co, coords, arg, standstill, vel, state, infection
#  75  __ append(Person(-)_object)で、
#  class_Person(引数)、実行で、戻り値 = Person_object(要素7)を、受取、
#  self.persons = [] に、戻り値を、追記
       
class Population:
    
    def __init__( self, number, comeback, p=None ):
        
        self.number = number
        self.R0 = (radius * (sqrt(2) * speed) * time_to_recover * self.number)/ (half ** 2)
        print("77_R0: %f" % self.R0)
        
        # 下記、persons にmenber_500を_append_記録
        self.persons = []
        self.p = p
        init_coords = []
        
        if self.p is not None:
            
            self.Re = (1.0 - self.p) * (1.0 + (sqrt(2) - 1) * self.p) * self.R0
            print("87_Re: %f" % self.Re,"\n")
            
        # menber_object(-7-)、生成___for-in[0,1,-,499],繰返,代入
        for i in range(self.number):
            
            if i == 0:
                
                # 最初の人間は感染「infected」者として初期条件を設定.
                # 初期座標は原点に設定.
                init_co = np.array([0.0, 0.0])
                
                print("98_配列_init_co=",init_co,"次元_shape=",init_co.shape," State.I.value = ",State.Infected.value,"\n")
                
                infected = Person(co=init_co,state=State.Infected,standstill=False)
                infected.infection_start = 0    # inf → 0
                
                self.persons.append(infected)
                
                print("105_リスト_persons[0]=",self.persons[0]," , ",len(self.persons),"\n")
                print("106_リスト_vars(persons[0])=",vars(self.persons[0]),"\n")
                
            else:
                #  ０～ｎの全ての座標と新座標を、for_で順に、比較し、重なり在りの場合は、
                # whileに戻り、random()で、_init_co、再作成し、for_で重なり確認
                
                k=0
                while True:
                    print("\n_114_while_",k,"\n")
                    init_co = np.array([uniform(-half, half), uniform(-half, half)])
                    appropriate = True
                    print("117_while_",k,"_init_co=",init_co,"  init_coords=",init_coords)
                    
                    #     座標重なり確認
                    j=0
                    for pre_init_co in init_coords:      # []
                        dist = np.linalg.norm(init_co - pre_init_co, ord=2)
                        j += 1
                        if dist <= 300 * radius:
                            appropriate = False    # 座標重なりで不適で、whileに戻る
                                                 
                    print("127_for-in_=",j)
                    
                    k += 1
                    
                    if appropriate: 
                        print("132_break_menber_i",i)
                        break
                        
                    if k==3:
                        print("136_break_menber_i",i)
                        break
                        
                        
                if self.p is None:
                    standstill = False        # 行動制限＝Noneは、全員、動かす
                    
                elif random() < self.p:        # 0 <random()< 1
                    standstill = True         # 2,random_に、停止
                    
                # その値が行動制限の割合以下,p%停止,以上(1-p)%動く  
                
                else:
                    standstill = False        # 3,動かす
                    
                self.persons.append(Person(co=init_co,
                                           state=State.Susceptible,
                                           standstill=standstill))
                
            init_coords.append(init_co)
            
            # リスト型は．a.append(b)、1点(b)のみ可能、2点以上不可
            print("\n 158_ menber_i =",i,"   len(init_coords)=",len(init_coords),"\n")
            
        self.times = np.array([0])
        self.zeros = np.array([0])
        self.I = np.array([1])
        self.S = np.array([self.number - 1])
        self.R = np.array([0])
        self.comeback = comeback
        
# 人口 3
if __name__ == "__main__":
    population = Population(number=7,
                            comeback=True,
                            p=0.8)
    
    print("173_Person.__name__ = ",Person.__name__)
#    population.simulate_sir()
```

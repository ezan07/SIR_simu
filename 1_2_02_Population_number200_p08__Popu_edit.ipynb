```python

import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from enum import Enum, auto

from random import uniform, random
from math import cos, sin, pi, inf, sqrt
import os

this_dir = "C:\\Users\\na366\\Desktop\\Python\\SIR"
radius = 0.075
dt = 0.1        # 時間刻み.dt = 0.01
half = 5
speed = 1
time_to_recover = 2    # 回復にかかる時間（回復率 γ の逆数）.
buff = 3        # バッファ. buff = 300

fig = plt.figure(figsize=plt.figaspect(1/1))
fig.set_facecolor('black')

#               auto_run_class_(Enum)

class State(Enum):
    Susceptible = auto()
    Infected = auto()
    R1=Recovered = auto()
    print("28_Stae.S,I,R=",Susceptible, int(Infected), int(R1),"\n")
    
class Person:
    
    # 92              [0,0]     2     Fulse
    
    def __init__( self, co, state, standstill ):
        
        self.co = co
        self.coords = np.array([self.co])          #  ([]) ⇒ [] 配列
        print("___40_Person_np.array([self.co])=",self.coords)
        #                                 [[0,0]] 2重[[]] は？
        
        self.arg = uniform(-pi, pi)                 # 乱数区間
        self.standstill = standstill               # 停止
        
        if self.standstill:
            #                    停止速[0,0] 
            vel = np.array([0.0, 0.0])
        else:
            #                      乱数（self.arg）
            vel = np.array([speed * cos(self.arg), speed * sin(self.arg)])

        self.vel = vel
        self.state = state
        self.infection_start = inf
        print("___56_Person.vel=", self.vel,",self.state=",self.state.value, "\n")

    def move( self ):
        
        # (self.standsrill =Ture =停止)→ (動かす)  line_231_call
        
        if self.standstill:
            self.vel = np.array([speed * cos(self.arg), speed * sin(self.arg)])
            #                 ( x ,y )arg乱数
            
            
            
class Population:
    
    # 人口    Population( 3     , Ture    ,0.8)
    
    def __init__( self, number, comeback, p=None ):
        
        self.number = number

        self.R0 = (radius * (sqrt(2) * speed) * time_to_recover * self.number)/ (half ** 2)

        print("78_R0: %f" % self.R0)
        self.persons = []            # 147_persons.append(Person(co-
        self.p = p
        init_coords = []             # 152_init_coords.append(init_co)
        
        if self.p is not None:
            
            self.Re = (1.0 - self.p) * (1.0 + (sqrt(2) - 1) * self.p) * self.R0
            print("86_Re: %f" % self.Re,"\n")
            
        # 総人口数分の人間を生成.     number=500,[0,1,---,499],繰返,代入
        for i in range(self.number):
            if i == 0:
                print("92_number_i=",i,"\n")
                # 最初の人間は感染「infected」者として初期条件を設定.
                # 初期座標は原点に設定.
                init_co = np.array([0.0, 0.0])

                print("96_init_co=",init_co,"shape=",init_co.shape," State.I.value = ",State.Infected.value,"\n")
                
                infected = Person(co=init_co,state=State.Infected,standstill=False)
                #                      [0,0]            2
                
                infected.infection_start = 0    # inf → 0
                
                print("103_if_i=0_追記前self.persons= ",",len=",len(self.persons),"\n",self.persons,"\n")    # 追記前
                self.persons.append(infected)
                #              []   append    None → ap1= None
                print("106_if_i=0_追記後self.persons=",",len=",len(self.persons),"\n",self.persons,"\n\n")
                # personsは、リスト型append
                
            else:
                print("\n\n_110_number_i=",i)
                # 原点の感染者一人[0]を除いた、499人[1,2,-,499]は、
                # ０～ｎの全ての散布座標を保存している_init_coords_(list型 と、
                #  random()で、作った、n+１座標_init_co_が、重ならない座標か
                #  ０～ｎの全ての座標と、for_で順に、比較し、重なり在りの場合は、
                # whileに戻り、random()で、_init_co、再作成し、for_で重なり確認
                
                k=0
                while True:
                    print("\n_116_while_",k,"\n")
                    init_co = np.array([uniform(-half, half), uniform(-half, half)])
                    appropriate = True
                    j=0
                    print("121_while_",k,"_init_co=",init_co,"  init_coords=",init_coords)

                    # init_coordsの要素を順に代入して、座標重なり確認
                    for pre_init_co in init_coords:      # []
                        print("123_while_",k,"_for",j,"_pre_init_co= ",pre_init_co)
                        dist = np.linalg.norm(init_co - pre_init_co, ord=2)
                        print("125_while_",k,"_for",j,"_L2ノルム_dist=",dist,"\n")
                        j += 1
                        if dist <= 300 * radius:
                            appropriate = False    # 座標重なりで不適で、whileに戻る
                    k += 1                          
                    if appropriate:
                        print("131_break_全ての散布座標が、重なし座標_",appropriate)
                        break
                    
                    if k==3:
                        break
                        
                        
                if self.p is None:
                    standstill = False        # 行動制限＝Noneは、全員、動かす
                elif random() < self.p:        # 0 <random()< 1
                    standstill = True         # 2,random_に、停止
                # その値が行動制限の割合以下,p%停止,以上(1-p)%動く    
                else:
                    standstill = False        # 3,動かす
                    
                print("146_else_number_",i,"_追記前self.persons",",len=",len(self.persons),"\n___",self.persons,"\n")    # 追記前
                self.persons.append(Person(co=init_co,
                                           state=State.Susceptible,
                                           standstill=standstill))
                print("150_else_number_",i,"_追記後self.persons",",len=",len(self.persons),"\n___",self.persons,"\n\n")    # 追記後
                
            init_coords.append(init_co)
            # リスト名．a.append(b)、1点のみ可能、2点以上不可
            print("150_init_coords,len=",len(init_coords),"→",init_coords)
            i += 1
            
        self.times = np.array([0])
        self.zeros = np.array([0])
        self.I = np.array([1])
        self.S = np.array([self.number - 1])
        self.R = np.array([0])
        self.comeback = comeback
        
# 人口 3
if __name__ == "__main__":
    population = Population(number=4,
                            comeback=True,
                            p=0.8)
    
    print("164_Person.__name__ = ",Person.__name__)
#    population.simulate_sir()
```
